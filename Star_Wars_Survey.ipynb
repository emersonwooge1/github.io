# Star Wars Survey

This project is based on a star wars survey with a total of 835 repsonses.  The goal of this project to examine which movies are the favorites and if there are any categorical variables which influence a particular favorite. This project will involve a good deal of data cleaning.

```python
import pandas as pd
import numpy as np
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")

#importing data, numpy and pandas
```

Based on the column names above, it appears questions were split out into multiple columns (for example have you seen star wars 1, 2, 3, 4, etc) where a bunch of the column were unamed and it just states the variable.  For example one column for have you seen Star wars 1, the header does not state this but the column values have "star wars 1" if they have seen it.

### Mapping Strings to Booleans

```python
yes_no = {"Yes": True,
         "No": False}

#creates a map to convert Yes/No variables to True or False
```

```python
star_wars["Have you seen any of the 6 films in the Star Wars franchise?"] = star_wars["Have you seen any of the 6 films in the Star Wars franchise?"].map(yes_no)

#Updates have you seen any star wars films from yes or no to true or false
```

```python
star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"] = star_wars["Do you consider yourself to be a fan of the Star Wars film franchise?"].map(yes_no)

#Updates do you consider yourself a star wars fan from yes or no to true or false
```

```python
star_wars.head()

#confirming values updated
```

### Cleaning Movies Seen Columns

```python
seen_1 = {"Star Wars: Episode I  The Phantom Menace": True,
          np.NaN: False}

seen_2 = {"Star Wars: Episode II  Attack of the Clones": True,
          np.NaN: False}

seen_3 = {"Star Wars: Episode III  Revenge of the Sith": True,
          np.NaN: False}

seen_4 = {"Star Wars: Episode IV  A New Hope": True,
          np.NaN: False}

seen_5 = {"Star Wars: Episode V The Empire Strikes Back": True,
          np.NaN: False}

seen_6 = {"Star Wars: Episode VI Return of the Jedi": True,
          np.NaN: False}

# creates mapping to replace movie name values with true or false
```

```python
star_wars[star_wars.columns[3]] = star_wars[star_wars.columns[3]].map(seen_1)
star_wars[star_wars.columns[4]] = star_wars[star_wars.columns[4]].map(seen_2)
star_wars[star_wars.columns[5]] = star_wars[star_wars.columns[5]].map(seen_3)
star_wars[star_wars.columns[6]] = star_wars[star_wars.columns[6]].map(seen_4)
star_wars[star_wars.columns[7]] = star_wars[star_wars.columns[7]].map(seen_5)
star_wars[star_wars.columns[8]] = star_wars[star_wars.columns[8]].map(seen_6)

# implements mapping created above. Could also use a for loop for this
```

```python
star_wars = star_wars.rename(columns={
    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
    "Unnamed: 4": "seen_2",
    "Unnamed: 5": "seen_3",
    "Unnamed: 6": "seen_4",
    "Unnamed: 7": "seen_5",
    "Unnamed: 8": "seen_6"
})

# renames columns as many are "unnamed" which is not easy to follow
```

```python
star_wars.head()

#confirming mapping worked correct
```

### Cleaning Movie Ranking Columns

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)

# updates the ranking columns to float
```

```python
star_wars = star_wars.rename(columns={
    "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.":"ranking_1",
    "Unnamed: 10": "ranking_2",
    "Unnamed: 11": "ranking_3",
    "Unnamed: 12": "ranking_4",
    "Unnamed: 13": "ranking_5",
    "Unnamed: 14": "ranking_6"
})

# updates ranking column names as we have "unnamed" again
```

### Rankings Analysis

```python
star_wars_fave_means = star_wars[star_wars.columns[9:15]].mean()

star_wars_fave_means #mean rankings of each movie
```

```python
%matplotlib inline
import matplotlib.pyplot as plt 

#imports mathplotlib
```

```python
star_wars_fave_means.plot.bar()

#plots mean rankings of each movie
```

Per the bar chart above we can conclude the older movies are more liked than the newer with start wars 3 being the favorite and star wars 5 being the least favorite

### Movies Seen Analysis

```python
star_wars_seen_sums = star_wars[star_wars.columns[3:9]].sum()

#sums total count seen by movie
```

```python
print(star_wars_seen_sums)
```

```python
star_wars_seen_sums.plot.bar()

#plots the total count seen by movie
```

It appears more people have seen the newer movies 5 and 6.  A good portion has also seen the first movie

### Female vs Male Analysis

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]

#creates two separate DFs for male and female
```

```python
male_fave_means = males[males.columns[9:15]].mean()
female_fave_means = females[females.columns[9:15]].mean()

#runs the mean on the favorite movie columns for both male and female
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1, 2, 1)
plt.title("Male Ranking")
male_fave_means.plot.bar()
plt.subplot(1, 2, 2)
plt.title("Female Ranking")
female_fave_means.plot.bar()
plt.show()

#plots the the average ranking by male and female
```

The average favorites have a similar shape above with the newer movies being more popular for the most part.  Females seemed to like the newer movies slightly more.

```python
star_wars_seen_male = males[star_wars.columns[3:9]].sum()
star_wars_seen_female = females[star_wars.columns[3:9]].sum()

#sums total count seen by movie
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1, 2, 1)
plt.title("Male Seen")
star_wars_seen_male.plot.bar()
plt.subplot(1, 2, 2)
plt.title("Female Seen")
star_wars_seen_female.plot.bar()
plt.show()

#plots total views by male and female
```

Per the chart above the trend is again very similar.  More females have seen the newer movies (5 and 6) vs older movies.
