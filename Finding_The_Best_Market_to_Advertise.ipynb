# Finding the Best Markets to Advertise In

In this project, we will assume we are working for an e-learning company which teaches various coding topics including: web and mobile development, data science, game development, etc.  Our goal is to figure out the most lucrative market to advertise in.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# importing libraries
```

### Data Exploration

```python
survey_data = pd.read_csv("2017-fCC-New-Coders-Survey-Data.csv", low_memory = False)

# importing data
```

Our data utilized above is from freeCodeCamp's 2017 New Coder Survey. freeCodeCamp is a free e-learning platform that offers courses on web development. Because they run a popular Medium publication (over 400,000 followers), their survey attracted new coders with varying interests (not only web development), which is ideal for the purpose of our analysis.

### Calculating Totals per Job Interest

```python
interests_no_nulls = survey_data.dropna(subset = ["JobRoleInterest"])

# removing data with non interest values
```

```python
interests_no_nulls.shape

# confirming values dropped
```

```python
job_interests = interests_no_nulls["JobRoleInterest"].value_counts()

# running value counts of job role interests
```

```python
job_interests_totals = sum(job_interests)

# creates a total sum of job role interests to obtain percentages in next step
```

```python
job_role = pd.DataFrame(data = {"count": job_interests,
                    }) # changes value counts to a pd dataframe

job_role["percentage"] = job_role["count"]/ job_interests_totals # adds a row for the overall percengage

job_role.sort_values(by=['percentage'], inplace=True, ascending=False) # sorts percentages in descending order
```

```python
job_role #view job interests counts
```

```python
job_role.reset_index(inplace=True)

# removes job role as index and creates new
```

```python
job_role = job_role.rename(columns = {'index':'job_interest'})

#creates column name for job interest
```

### Comparing Totals for one Job Interest vs Multiple

```python
job_role["contains_comma"] = job_role['job_interest'].str.contains(',')

#creates a boolean column if the job interest value has a comma
```

```python
single_interest = job_role[job_role["contains_comma"] == True]
multiple_interest = job_role[job_role["contains_comma"] == False]

# creates two separate dataframes to determine totals by single role vs multiple roles
```

```python
print(single_interest["count"].sum()/ job_interests_totals)
print(multiple_interest["count"].sum()/job_interests_totals)

# calculates the percentage of single role intresests vs multiple
```

### Determining the Best Market to Advertise

Per the counts above, the US and India would be the biggest markets.  The US is by far the largest Country in the sample.

### Average Amount Spent per Month by Country

```python
interests_no_nulls['MonthsProgramming'].replace(0,1, inplace = True)
interests_no_nulls["monthly_spent"] = interests_no_nulls["MoneyForLearning"]/ interests_no_nulls["MonthsProgramming"]

# assumes blanks have at least one month of programming creates a money spend per month column
```

```python
interests_no_nulls = interests_no_nulls[interests_no_nulls['monthly_spent'].notnull()]

# removes any null values calculated
```

```python
countries_mean = interests_no_nulls.groupby('CountryLive').mean()

# Groups the data by country lived in and creates means for quantitiative variables
```

```python
countries_mean["monthly_spent"].sort_values(ascending = False)

# filters on grouped mean table on monthly spent
```

```python
interests_no_nulls = interests_no_nulls[interests_no_nulls['CountryLive'].notnull()]
top_countries = interests_no_nulls[interests_no_nulls["CountryLive"].str.contains("United States of America|India|United Kingdom|Canada")]

# filters on our top markets based on actual users
```

```python
sns.boxplot(y = 'monthly_spent', x = 'CountryLive',
            data = top_countries)

plt.xticks(rotation = 45)

# creates box plot of top markets
```

```python
top_countries = top_countries[top_countries["monthly_spent"] < 30000]

# per the chart above we have some extreme outliers in the US.  Computing means again without these
```

```python
sns.boxplot(y = 'monthly_spent', x = 'CountryLive',
            data = top_countries)

plt.xticks(rotation = 45)

# creates box plot of top markets without extreme US outliers
```

```python
top_countries_mean = top_countries.groupby('CountryLive').mean()

# running means again without outliers
```

```python
top_countries_mean["monthly_spent"].sort_values(ascending = False)

# means excluding extreme outliers
```

### Conclusions

Removing the significant outliers brought down the mean spent in the US by over $40 which should make our sample more realistic. The US is clearly our best market to advertise in as it has the highest average amount spent per month and by far the largest sample population. India and Canada have a fairly close average amount spent per month but India's sample population over 200 PN higher than Canada's, making it a better market to target.

Now that we've determined the US and India are the best markets to advertise in, we also want to target the most popular subjects.  Our sample found that 86% of all users wanted courses on web or mobile development.  I would recommend targeting these skills to take advantage of the large sample seeking these services.
